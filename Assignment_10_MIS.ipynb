```python
price_sugarshake = 1.50
price_heartstopper = 2.00
price_diabeauty = 2.50

qty_sugarshake = 2
qty_heartstopper = 1
qty_diabeauty = 3

subtotal = price_sugarshake * qty_sugarshake + price_heartstopper * qty_heartstopper + price_diabeauty * qty_diabeauty

total_donuts = qty_sugarshake + qty_heartstopper + qty_diabeauty

if total_donuts >= 12:
    discount = subtotal * 0.05
    subtotal = subtotal - discount
else:
    discount = 0

tax = subtotal * 0.065

total = subtotal + tax

print("Your subtotal is:", subtotal)
print("Your tax is:", tax)
print("Your total is:", total)
```
